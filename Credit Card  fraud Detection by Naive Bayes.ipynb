```python
### CREATE TESTING AND TRAINING DATASET
```

# lets normalize amount coulmn using feature extraction

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
creditcard_df['Amountnorm'] =sc.fit_transform(creditcard_df['Amount'].values.reshape(-1, 1))
```

```python
creditcard_df['Amountnorm']
```

```python
creditcard_df.drop(['Amount'], axis = 1)
```

```python
### now our amount column is normlaized
```

```python
## lets drop target label
```

```python
X =  creditcard_df.drop(['Class'], axis = 1)
```

```python
Y = creditcard_df['Class']
```

```python
X 
```

```python
Y
```

```python
## Dive data into training and testing
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)
```

```python
#### Training the model
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
## we gonna use gaussian classifier
```

```python
from sklearn.naive_bayes import GaussianNB
NB_classifier = GaussianNB()
```

```python
NB_classifier.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_predict_train = NB_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
cm = confusion_matrix(Y_train, y_predict_train)
```

```python
sns.heatmap(cm , annot = True)
```

```python
# predict Testing results
```

```python
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(Y_test, y_predict_test)
sns.heatmap(cm , annot = True)
```

```python
print(classification_report(Y_test, y_predict_test))
```
